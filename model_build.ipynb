###  Keeping male and female separate

```python
f_data = data.loc[data['gender'] == 1]      #Females
```

```python
m_data = data.loc[data['gender'] == 0]    #Males
```

```python
def similarity_score(x, y):
    score = 0.0
    for i in range(len(x)):
        if x[i] == y[i]: score+=1
    return(score)
```

```python
similarity_score(m_data.iloc[0], m_data.iloc[200])
```

```python
m_data.shape
```

```python
simi_mat = np.zeros((m_data.shape[0], m_data.shape[0]))
```

```python
for i in range(len(m_data)):
    for j in range(i, len(m_data)):
        simi_mat[i,j] = similarity_score(m_data.iloc[i], m_data.iloc[j])
```

```python
simi_mat
```

```python
top_10_peruser = np.zeros((len(simi_mat), 10))
for i, row in enumerate(simi_mat):
    top_10_peruser[i, :] = np.argsort(-row)[1:11]
```

```python
top_10_peruser
```

```python
for j in top_10_peruser[0]:
    print(m_data.loc[j])
```
